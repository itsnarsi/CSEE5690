```python
# import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
import numpy as np
import pandas as pd
```

```python
F = '/home/narsi/Documents/CSEE5690/Final_project/mapping_all_ips.csv'
T = pd.read_csv(F)
log_size = len(T)
```

```python
T = T[T['count']>100]
```

```python
T.index = range(len(T))
```

```python
T
```

```python
sum_ = np.sum(T['count'])
flow_ = []
labels_ = []
for i in range(0,len(T)):
    count = T['count'][i]
    flow_.append(count/1000)
    flow_.append(-1*count/1000)
    labels_.append(T['has'][i])
    labels_.append(T['prefers'][i])
```

```python
str_ = ''
for i in range(0,len(T)): #len(T)
    str_+= '[\''
    str_+= str(T['has'][i])
    str_+= '\''
    str_+= ','
    str_+= '\''
    str_+= str(T['prefers'][i])
    str_+= '\''
    str_+= ','
    str_+= str(T['count'][i])
    str_+= '],'
```

```python
text_file = open("Output.txt", "w")
text_file.write(str_)
text_file.close()
```

```python
fig = plt.figure(figsize=(50, 50))
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[],
                     title="IP mapping")
sankey = Sankey(ax = ax)
sankey.add(flows=flow_,orientations=[0]*len(flow_),
       labels=labels_)
sankey.finish()
```
