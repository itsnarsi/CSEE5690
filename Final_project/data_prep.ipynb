```python
%matplotlib inline
import pandas as pd
import numpy as np
import pickle
import time
import scipy
import datetime
import decimal
from multiprocessing import Pool

F = '/home/narsi/Documents/CSEE5690_data/PaloAlto/UMKC-Palo-5Nov-to-9Nov.csv'
T = pd.read_csv(F)
log_size = len(T)
```

```python
ALL_IPS = np.unique(np.concatenate((T['Source address'].unique(), T['Destination address'].unique()),axis=0))
```

```python
MAP_CSV = pd.DataFrame({'has':[],'prefers':[],'count':[]})
nn = len(ALL_IPS)
with Pool(processes=8) as pool:

    for i in ALL_IPS:
        x = pd.Series.value_counts(T['Destination address'][(T['Source address'] == i)])
        if len(x) == 0:
            MAP_CSV = MAP_CSV.append(pd.DataFrame({'has':[i] * nn,'prefers':ALL_IPS,'count':[0] * nn}),ignore_index=True)
        else:
            x = int(x[0])
            counter = 0
            for j in ALL_IPS:
                if x > 0:
                    c = pd.Series.value_counts(T['Destination address'][(T['Source address'] == i) & (T['Destination address'] == j)])                
                    if len(c) == 0:
                        c = 0
                    else:
                        c = int(c[0])
                    x = x - c
                    counter = counter + 1
                    MAP_CSV = MAP_CSV.append(pd.DataFrame({'has':i,'prefers':j,'count':[c]}),ignore_index=True)
                else:
                    c = 0
                    X = ALL_IPS[counter:]
                    MAP_CSV = MAP_CSV.append(pd.DataFrame({'has':[i]*len(X),'prefers':X,'count':[c]*len(X)}),ignore_index=True)
                    break

        print(i)
```
