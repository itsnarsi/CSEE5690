```python
%matplotlib inline
import pandas as pd
import numpy as np
import pickle
import time
import scipy
import datetime
import decimal
from multiprocessing import Pool

F = '/home/narsi/Documents/CSEE5690_data/PaloAlto/UMKC-Palo-5Nov-to-9Nov.csv'
T = pd.read_csv(F)
log_size = len(T)
```

```python
ALL_IPS = np.unique(np.concatenate((T['Source address'].unique(), T['Destination address'].unique()),axis=0))
```

```python
MAP_CSV = pd.DataFrame({'has':[],'prefers':[],'count':[]})
for i in ALL_IPS:
    U = T['Destination address'][(T['Source address'] == i)].unique()
    for j in U:
        c = pd.Series.value_counts(T['Destination address'][(T['Source address'] == i) & (T['Destination address'] == j)])                
        MAP_CSV = MAP_CSV.append(pd.DataFrame({'has':i,'prefers':j,'count':[int(c[0])]}),ignore_index=True)
    V = list(set(ALL_IPS) - set(U))
    MAP_CSV = MAP_CSV.append(pd.DataFrame({'has':[i]*len(V),'prefers':V,'count':[0]*len(V)}),ignore_index=True)
    print(i)
```

```python
MAP_CSV.to_csv('mapping_all_ips.csv',',', index=False)
```
