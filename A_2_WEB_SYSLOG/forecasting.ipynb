```python
data_2_fore = unique_vistors
name = 'unique vistors'
window = 7
mov_avg = np.empty(len(data_2_fore) - window)
mov_lin_reg = np.empty(len(data_2_fore) - window)
t = np.arange(window)
counter = 0
for i in range(window,len(data_2_fore)):
    X = data_2_fore[i - window:i]
    mov_avg[counter] = np.sum(X) * 1.0 / window
    (ar,br) = scipy.polyfit(t,X,1)
    mov_lin_reg[counter] = scipy.polyval([ar,br],window)
    counter += 1

mov_avg_err = (np.sum(np.absolute(data_2_fore[window:]-mov_avg))/len(mov_avg))
mov_lin_reg_err = (np.sum(np.absolute(data_2_fore[window:]-mov_lin_reg))/len(mov_lin_reg))

expo_smooth = np.empty(len(data_2_fore))
expo_smooth[0] = data_2_fore[0]

alpha = 0.8

for i in range(1,len(data_2_fore)):
    expo_smooth[i] = expo_smooth[i-1] + (alpha) * (data_2_fore[i-1] - expo_smooth[i-1])

expo_smooth_err = (np.sum(np.absolute(data_2_fore-expo_smooth))/len(expo_smooth))


def datetime_1(x):
    y = []
    for i in range(len(x)):
        y.append(datetime.strptime(x[i],'%Y-%m-%d'))
    return np.array(y, dtype=np.datetime64)


line_type = ['solid', 'dashed', 'dotted']

p2 = figure(x_axis_type="datetime",title=name,width = 1000,height = 700)
p2.xaxis.axis_label = 'Date'
p2.yaxis.axis_label = 'Count'

p2.line(datetime_1(dates),data_2_fore,color = Set1_5[1], line_width=2,legend='Original')
p2.line(datetime_1(dates[window:]),mov_avg,color = Set1_5[0], line_width=3,line_dash = 'solid',legend='Moving Average '+str('%.2f' %mov_avg_err))
p2.line(datetime_1(dates[window:]),mov_lin_reg,color = Set1_5[2], line_width=3,line_dash = 'dashed',legend='Moving Linear Regression '+str('%.2f' %mov_lin_reg_err))
p2.line(datetime_1(dates),expo_smooth,color = Set1_5[3], line_width=3,line_dash = 'dotted',legend='Exponential Smoothing '+str('%.2f' %expo_smooth_err))
p2.legend.location = "top_left"
show(p2)
```
